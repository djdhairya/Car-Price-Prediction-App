```python
df=df.sort_values(by='Selling_Price',ascending=False).reset_index()
```

```python
#remove outliers
df=df[3:].reset_index()
df.head()
```

```python
df.drop(['level_0','index'],axis=1,inplace=True)
df
```

```python
df.info()
```

```python
object=['Car_Name','Fuel_Type','Seller_Type','Transmission']
for names in object:
    count=df[names].value_counts()
    print(str(names))
    print(str(count))
    print("---------")
```

```python
target_encode_column='Car_Name'
encoder=ce.TargetEncoder(cols=[target_encode_column])
df_encoded=encoder.fit_transform(df,df['Selling_Price'])
```

```python
df_encoded
```
